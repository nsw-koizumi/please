```python
from pyspark.sql import SparkSession

# Sparkセッションの作成
spark = SparkSession.builder \
    .appName("CSV to Bronze Table") \
    .getOrCreate()

# CSVファイルのパスをリストに格納
csv_file = ["/FileStore/tables/product_data.csv"]

# 各CSVファイルを読み込み、DataFrameを作成
df = spark.read.csv(csv_file, header=True, inferSchema=True)

# 一時ビューを作成
df.createOrReplaceTempView("TEMP_MST_PRODUCT")

```

```python
# 新しいテーブルを作成
table_name = "test_wsc.MST_PRODUCT"
if not spark.catalog.tableExists(table_name):
    spark.sql("""
    CREATE TABLE test_wsc.MST_PRODUCT (
        COMPANY_CODE STRING NOT NULL,
        PRODUCT_CODE_LOCAL STRING NOT NULL,
        PRODUCT_NAME_LOCAL STRING,
        ZOFF_COLOR1_LOCAL STRING,
        ZOFF_COLOR2_LOCAL STRING,
        MODEL_NO_LOCAL STRING,
        PRODUCT_CODE STRING,
        PRODUCT_NAME STRING,
        ZOFF_COLOR1 STRING,
        ZOFF_COLOR2 STRING,
        MODEL_NO STRING,
        PRODUCT_CODE_FOR_PIC STRING,
        RELATED_MODEL_NO STRING,
        LARGE_CLASS_CODE STRING,
        LARGE_CLASS_NAME STRING,
        MIDDLE_CLASS_CODE STRING,
        MIDDLE_CLASS_NAME STRING,
        SMALL_CLASS_CODE STRING,
        SMALL_CLASS_NAME STRING,
        MATERIAL STRING,
        LENS_SAHPE_CODE STRING,
        LENS_SAHPE_NAME STRING,
        FLAGSHIP_PRODUCT STRING,
        SELLING_PRICE DECIMAL(8, 2),
        TRADE_NAME STRING,
        BRAND_CODE STRING,
        BRAND_NAME STRING,
        PO_CATEGORY_CODE STRING,
        PO_CATEGORY_NAME STRING,
        FIRST_RELEASE_DATE TIMESTAMP,
        RELEASE_DATE TIMESTAMP,
        RELEASE_DATE_FOR_MARCHANDISE TIMESTAMP,
        SHAPE1_CODE STRING,
        SHAPE1_NAME STRING,
        SHAPE2_CODE STRING,
        SHAPE2_NAME STRING,
        FRAME_HEIGHT DECIMAL(8, 2),
        LENS_WIDTH DECIMAL(8, 2),
        TARGET_CODE STRING,
        TARGET_NAME STRING,
        REPEAT_COUNT DECIMAL(8, 2),
        ORDER_RANK STRING,
        DISCONTINUED_FLAG STRING,
        PRODUCT_CATEGORY_MATERIAL STRING,
        PRODUCT_CATEGORY_LENS_SHAPE STRING,
        PRODUCT_CATEGORY_MIDDLE_CLASS STRING,
        PC_LENS_MIDDLE_CLASS STRING,
        PROGRESSIVE_LENS_MIDDLE_CLASS STRING,
        FIRST_SELLING_PRICE DECIMAL(8, 2),
        UNIT_IM_COST_PRICE DECIMAL(8, 2),
        UNIT_NET_PRICE_A DECIMAL(8, 2),
        UNIT_NET_PRICE_B DECIMAL(8, 2),
        UNIT_NET_PRICE_C DECIMAL(8, 2),
        UNIT_NET_PRICE_D DECIMAL(8, 2),
        UNIT_NET_PRICE_E DECIMAL(8, 2),
        UNIT_SELLING_PRICE_A DECIMAL(8, 2),
        UNIT_SELLING_PRICE_B DECIMAL(8, 2),
        UNIT_SELLING_PRICE_C DECIMAL(8, 2),
        UNIT_SELLING_PRICE_D DECIMAL(8, 2),
        UNIT_SELLING_PRICE_E DECIMAL(8, 2),
        CN_FLAGSHIP_PRODUCT_CLASS STRING,
        CN_SKINNY_CLASS STRING,
        CN_SMART_CLASS STRING,
        SALES_FORM STRING,
        CN_LENS_CLASS STRING,
        CN_LENS_CATEGORY STRING,
        ACCOUT_PATTERN_CODE STRING,
        ACCOUT_PATTERN_NAME STRING,
        ACCOUNT_TITLE_CODE STRING,
        ACCOUNT_TITLE_NAME STRING,
        PRODUCT_CLASS STRING,
        SUBSIDIARY_ACCOUNT_CODE STRING,
        SUBSIDIARY_ACCOUNT_NAME STRING,
        MANUFUCTURER_CODE STRING,
        FACTORY_CODE STRING,
        INVENTORY_CATEGORY_CODE STRING,
        INVENTORY_CATEGORY_NAME STRING,
        LARGE_CATEGORY_CODE STRING,
        LARGE_CATEGORY_NAME STRING,
        MIDDLE_CATEGORY_CODE STRING,
        MIDDLE_CATEGORY_NAME STRING,
        SMALL_CATEGORY_CODE STRING,
        SMALL_CATEGORY_NAME STRING,
        ACQUISITION_DATE TIMESTAMP,
        PRODUCT_CD_WITH_CMPNY STRING NOT NULL,
        JAN_CODE STRING,
        PRIMARY KEY (COMPANY_CODE, PRODUCT_CODE_LOCAL, PRODUCT_CD_WITH_CMPNY)
    ) USING DELTA
    """)
```
